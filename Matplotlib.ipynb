# Very important library for data visualization 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10)
y= x * 2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show();
```

```python
# it is possible to add titles 
plt.plot(x,y)
plt.title(' X & y plot')
plt.xlabel(' the x axis')
plt.ylabel('the y axis')
plt.show();
```

```python
# to give limits for x-axis &/or y-axis 
plt.plot(x,y)
plt.xlim(0,8)
plt.ylim(0,16)
plt.show();
```

```python
# To save the figure acommend is use ( plt.savefig('path & name . extension '))
```

```python
# To controal the size and resolution of the figure 
plt.figure(figsize=(8,6), dpi=150)
plt.plot(x,y)
plt.show()
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
fig =plt.figure()
axes=fig.add_axes([0,0,1,1])
plt.plot(x,y)
plt.show();
```

```python
# can plot another plot inside !
```

```python
fig =plt.figure()
axes_big=fig.add_axes([0,0,1,1])
axes_big.plot(a,b)

axes_small = fig.add_axes([0.15,0.3,0.5,0.5])
axes_small.plot(x,y)


plt.show();
```

```python
# can add lables as well as limits 
fig =plt.figure()
axes_big=fig.add_axes([0,0,1,1])
axes_big.plot(a,b)

axes_small = fig.add_axes([0.15,0.3,0.5,0.5])
axes_small.plot(x,y)

axes_small.set_title('Increasing')
axes_small.set_xlabel('XX')
axes_small.set_ylabel('YY')

axes_small.set_xlim(1,20)
axes_small.set_ylim(0,80)

plt.show();
```

```python
# possible to make the plots in another shape like 
fig =plt.figure()
axes_big=fig.add_axes([0,0,1,1])
axes_big.plot(a,b)

axes_small = fig.add_axes([1,1,2,2])
axes_small.plot(x,y)

axes_small.set_title('Increasing')
axes_small.set_xlabel('XX')
axes_small.set_ylabel('YY')

axes_small.set_xlim(1,20)
axes_small.set_ylim(0,80)

plt.show();
```

```python
# It possible to drow more than one function with-in same plot
fig = plt.figure()
axes0 = fig.add_axes([0,0,1,1])
axes0.plot(x,y, label=('X & Y'))
axes0.plot(a,b, label=('A & B'))

plt.show()
```

# Subplots 

```python
fig,axes1 = plt.subplots(nrows=1 , ncols = 2)
axes1[0].plot(x,y)
axes1[1].plot(a,b)
plt.show()
```

```python
fig,axes2 = plt.subplots(nrows=2,ncols=2)
axes2[0,0].plot(a,b)
axes2[0,1].plot(x,y)
axes2[1,0].plot(a,a*2)
axes2[1,1].plot(y,x)

axes2[1,0].set_title(' a & x')
fig.suptitle(' test ')
plt.show()
```

```python
# to adjust space between the plots
# ( fig.tight_layout()) OR ( fig.subplots_adjest(wspace=0.9,hspace=0.9))
```

```python
fig,axes2 = plt.subplots(nrows=2,ncols=2)
axes2[0,0].plot(a,b)
axes2[0,1].plot(x,y)
axes2[1,0].plot(a,a*2)
axes2[1,1].plot(y,x)

axes2[1,0].set_title(' a & x')
fig.suptitle(' test ')

fig.subplots_adjust(wspace=0.9,hspace=0.9)

plt.show()
```

```python
fig,axes2 = plt.subplots(nrows=2,ncols=2)
axes2[0,0].plot(a,b)
axes2[0,1].plot(x,y)
axes2[1,0].plot(a,a*2)
axes2[1,1].plot(y,x)

axes2[1,0].set_title(' a & x')
fig.suptitle(' test ')

fig.tight_layout()

plt.show()
```
